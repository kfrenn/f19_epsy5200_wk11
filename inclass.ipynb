Challenge 2: Find which columns have missing data (and how many missing data)

```python
# challenge 2 code here
mov.isnull().sum()
mov2=mov.drop(['homepage','tagline'], axis=1) #axis 1 refers to columns, axis0 referes to rows

mov2.isnull().sum()
mov2.head()


```

```python
mov2.isnull().sum()
```

```python
mov3=mov2[(mov2.budget>0)&(mov2.revenue>0)]

```

```python
mov3.describe().transpose()
```

```python
mov3.shape
```

```python
sns.regplot(mov3.budget, mov3.revenue)
```

```python
rev_budg=smf.ols(formula='revenue ~ budget',data=mov3).fit()
```

```python
rev_budg
```

```python
rev_budg.params
```

```python
rev_budg.rsquared
rev_budg.pvalues
```

```python
rev_budg.conf_int()
```

```python
rev_budg.summary()
```

```python
output=pd.DataFrame({'estimate':rev_budg.params, 'lowCI':rev_budg.conf_int()[0],'highCI':rev_budg.conf_int()[1], 'pVal':rev_budg.pvalues}
            )
output
```

```python
output.to_csv('regression_table.csv')
```
